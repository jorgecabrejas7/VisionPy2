```python
import numpy as np
import plotly.graph_objects as go

# Assuming your 3D labeled array has dimensions (350, 40, 40)
labels = np.zeros((350, 40, 40))  # Replace this with your actual data

labels[-1] = np.ones((40, 40))  # Add some non-zero labels for demonstration purposes

# Generate a meshgrid for the (z, x, y) coordinates matching your volume shape
z, x, y = np.mgrid[0:350, 0:40, 0:40]

# Filter out points where the label is 0
non_zero_mask = labels != 0

# Extract the coordinates and labels of non-zero points
z_points = z[non_zero_mask]
x_points = x[non_zero_mask]
y_points = y[non_zero_mask]
label_points = labels[non_zero_mask]

# Create a 3D scatter plot using plotly
fig = go.Figure(
    data=[
        go.Scatter3d(
            x=x_points,
            y=y_points,
            z=z_points,
            mode="markers",
            marker=dict(
                size=3,
                color=label_points,
                colorscale="Viridis",
                colorbar=dict(title="Label"),
            ),
        )
    ]
)

# Set plot title and axis labels
fig.update_layout(
    title="3D Labeled Array (Non-zero Labels Only)",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis",
        zaxis=dict(autorange="reversed"),
    ),
)  # Invert Z axis

# Show the plot
fig.show()
```
